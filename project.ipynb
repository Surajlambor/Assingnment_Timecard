```python
import pandas as pd
file_path = 'Assignment_Timecard.csv'

df = pd.read_csv(file_path)

print(df.head())

    
```

```python

df.drop(['Unnamed: 9', 'Unnamed: 10'], axis=1, inplace=True)
df.dropna(inplace=True)
df.isnull().sum()
```

```python
from datetime import datetime 


df['Work Date'] = pd.to_datetime(df['Time']).dt.date

df_sorted = df.sort_values(by=['Employee Name', 'Work Date'])

def has_seven_consecutive_days(dates):

  dates = sorted(set(dates)) 

  for i in range(len(dates)):
#check there are atleast 7 days
    if i + 6 < len(dates):

      if (dates[i + 6] - dates[i]).days == 6:

        return True

  return False
#above function is apply for each employee workday
grouped = df_sorted.groupby('Employee Name')['Work Date'].apply(has_seven_consecutive_days)

employees_with_7_consecutive_days = grouped[grouped].index.tolist()
#Display the names of employees who worked 7 consecutive days
print('Employees who worked 7 consecutive days:', employees_with_7_consecutive_days)
```

```python
# Convert Time and Time Out columns to datetime
for col in ['Time', 'Time Out']:
    df[col] = pd.to_datetime(df[col])

#sort by Employee Name and Time
sorted_df = df.sort_values(by=['Employee Name', 'Time'])

#calculate the time difference between consecutive shifts
sorted_df['Time Difference'] = sorted_df.groupby('Employee Name')['Time'].diff().shift(-1).dt.total_seconds().div(3600)

# Filter employees with less than 10 hours but more than 1 hour between shifts
filtered_df = sorted_df[(sorted_df['Time Difference'] > 1) & (sorted_df['Time Difference'] < 10)]

# Print the name and position of these employees
for index, row in filtered_df.iterrows():
    print(f"Employee Name: {row['Employee Name']}, Position ID: {row['Position ID']}")

```

```python
# Calculate the total hours worked in a single shift
# Convert Time to total hours

def convert_to_hours(time_str):
    if pd.isnull(time_str):
        return 0
    hours, minutes = map(int, time_str.split(':'))
    return hours + minutes / 60


df['Total Hours'] = df['Timecard Hours (as Time)'].apply(convert_to_hours)

# Filter employees who have worked more than 14 hours in a single shift
long_shifts_df = df[df['Total Hours'] > 14]

# Print the name and position of these employees
for index, row in long_shifts_df.iterrows():
    print(f"Employee Name: {row['Employee Name']}, Position ID: {row['Position ID']}")
```
